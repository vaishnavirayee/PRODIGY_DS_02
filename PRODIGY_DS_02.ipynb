```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = pd.read_csv('titanic.csv')
```

```python

print(titanic_data.head())


print(titanic_data.info())


print(titanic_data.isnull().sum())
```

```python
# Summary statistics for numerical features
print(titanic_data.describe())
```

```python

sns.histplot(titanic_data['Age'], kde=True)
plt.title('Distribution of Age')
plt.show()


sns.histplot(titanic_data['Fare'], kde=True)
plt.title('Distribution of Fare')
plt.show()
```

```python

sns.countplot(x='Survived', data=titanic_data)
plt.title('Survival Count')
plt.show()


sns.countplot(x='Pclass', data=titanic_data)
plt.title('Passenger Class Count')
plt.show()


sns.countplot(x='Embarked', data=titanic_data)
plt.title('Embarkation Count')
plt.show()
```

```python

sns.barplot(x='Sex', y='Survived', data=titanic_data)
plt.title('Survival Rate by Sex')
plt.show()


sns.barplot(x='Pclass', y='Survived', data=titanic_data)
plt.title('Survival Rate by Class')
plt.show()

sns.barplot(x='Embarked', y='Survived', data=titanic_data)
plt.title('Survival Rate by Embarkation')
plt.show()


sns.pairplot(titanic_data[['Survived', 'Age', 'Fare', 'Pclass']])
plt.show()
```

```python

titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1


sns.barplot(x='FamilySize', y='Survived', data=titanic_data)
plt.title('Survival Rate by Family Size')
plt.show()
```
